```python
import pandas as pd
import pymysql
```

```python
df = pd.read_csv("Agriculrural_data.csv")
```

```python
myconnection = pymysql.connect(host = 'localhost',user='root',passwd = '1989')

cursor = myconnection.cursor()

cursor.execute("create database agricultural_db")
```

```python
cursor.execute("use agricultural_db")
```

```python
cursor.execute("""CREATE TABLE IF NOT EXISTS agriculture(
id INT AUTO_INCREMENT PRIMARY KEY,
Year INT,
State_Name VARCHAR(100),
Dist_Name VARCHAR(100),
RICE_AREA_1000_ha DECIMAL(12,2),
RICE_PRODUCTION_1000_tons DECIMAL(12,2),
RICE_YIELD_Kg_per_ha DECIMAL(12,2),
WHEAT_AREA_1000_ha DECIMAL(12,2),
WHEAT_PRODUCTION_1000_tons DECIMAL(12,2),
WHEAT_YIELD_Kg_per_ha DECIMAL(12,2),
KHARIF_SORGHUM_AREA_1000_ha DECIMAL(12,2),
KHARIF_SORGHUM_PRODUCTION_1000_tons DECIMAL(12,2),
KHARIF_SORGHUM_YIELD_Kg_per_ha DECIMAL(12,2),
RABI_SORGHUM_AREA_1000_ha DECIMAL(12,2),
RABI_SORGHUM_PRODUCTION_1000_tons DECIMAL(12,2),
RABI_SORGHUM_YIELD_Kg_per_ha DECIMAL(12,2),
PEARL_MILLET_AREA_1000_ha DECIMAL(12,2),
PEARL_MILLET_PRODUCTION_1000_tons DECIMAL(12,2),
PEARL_MILLET_YIELD_Kg_per_ha DECIMAL(12,2),
MAIZE_AREA_1000_ha DECIMAL(12,2),
MAIZE_PRODUCTION_1000_tons DECIMAL(12,2),
MAIZE_YIELD_Kg_per_ha DECIMAL(12,2),
FINGER_MILLET_AREA_1000_ha DECIMAL(12,2),
FINGER_MILLET_PRODUCTION_1000_tons DECIMAL(12,2),
FINGER_MILLET_YIELD_Kg_per_ha DECIMAL(12,2),
GROUNDNUT_AREA_1000_ha DECIMAL(12,2),
GROUNDNUT_PRODUCTION_1000_tons DECIMAL(12,2),
GROUNDNUT_YIELD_Kg_per_ha DECIMAL(12,2),
SUNFLOWER_AREA_1000_ha DECIMAL(12,2),
SUNFLOWER_PRODUCTION_1000_tons DECIMAL(12,2),
SUNFLOWER_YIELD_Kg_per_ha DECIMAL(12,2),
SOYABEAN_AREA_1000_ha DECIMAL(12,2),
SOYABEAN_PRODUCTION_1000_tons DECIMAL(12,2),
SOYABEAN_YIELD_Kg_per_ha DECIMAL(12,2),
OILSEEDS_AREA_1000_ha DECIMAL(12,2),
OILSEEDS_PRODUCTION_1000_tons DECIMAL(12,2),
OILSEEDS_YIELD_Kg_per_ha DECIMAL(12,2),
SUGARCANE_AREA_1000_ha DECIMAL(12,2),
SUGARCANE_PRODUCTION_1000_tons DECIMAL(12,2),
SUGARCANE_YIELD_Kg_per_ha DECIMAL(12,2),
COTTON_AREA_1000_ha DECIMAL(12,2),
COTTON_PRODUCTION_1000_tons DECIMAL(12,2),
COTTON_YIELD_Kg_per_ha DECIMAL(12,2)
)""")
print("Table created Successfully")

```
