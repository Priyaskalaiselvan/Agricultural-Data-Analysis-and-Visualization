```python
#Changing the negative values to nan
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols].applymap(lambda x: np.nan if x < 0 else x)
```

```python
def fix_crop_values(df, crop_name):
    area = f"{crop_name}_AREA_1000_ha"
    prod = f"{crop_name}_PRODUCTION_1000_tons"
    yld = f"{crop_name}_YIELD_Kg_per_ha"
    
    # Case 1: area > 0 & production > 0 → calculate yield
    mask1 = (df[area] > 0) & (df[prod] > 0)
    df.loc[mask1, yld] = (df.loc[mask1, prod] * 1000) / df.loc[mask1, area]
    
    # Case 2: area = 0 & production = 0 → yield = 0
    mask2 = (df[area] == 0) & (df[prod] == 0)
    df.loc[mask2, yld] = 0
    
    # Case 3: area = 0 & production > 0 → impossible → set all to NaN
    mask3 = (df[area] == 0) & (df[prod] > 0)
    df.loc[mask3, [area, prod, yld]] = np.nan
    
    # Case 4: area > 0 & production = 0 → crop failure/missing → yield = 0
    mask4 = (df[area] > 0) & (df[prod] == 0)
    df.loc[mask4, yld] = 0
    
    return df
```

```python
crops = ["RICE","WHEAT","KHARIF_SORGHUM","RABI_SORGHUM","PEARL_MILLET","MAIZE","FINGER_MILLET","GROUNDNUT","SUNFLOWER","SOYABEAN","OILSEEDS","SUGARCANE","COTTON"]
for crop in crops:
    df = fix_crop_values(df, crop)
```

```python
#Filling median value
def fill_statewise_median(df, crop):
    area = f"{crop}_AREA_1000_ha"
    prod = f"{crop}_PRODUCTION_1000_tons"
    yld  = f"{crop}_YIELD_Kg_per_ha"
    
    # fill area, production, yield with median grouped by state
    df[area] = df.groupby("State_Name")[area].transform(
        lambda x: x.fillna(x.median())
    )
    df[prod] = df.groupby("State_Name")[prod].transform(
        lambda x: x.fillna(x.median())
    )
    df[yld] = df.groupby("State_Name")[yld].transform(
        lambda x: x.fillna(x.median())
    )
    
    return df
```

```python
crops = ["RICE","WHEAT","KHARIF_SORGHUM","RABI_SORGHUM","PEARL_MILLET","MAIZE","FINGER_MILLET","GROUNDNUT","SUNFLOWER","SOYABEAN","OILSEEDS","SUGARCANE","COTTON"]
for crop in crops:
    df = fill_statewise_median(df, crop)
```

```python
def recalc_yield(df, crop):
    area = f"{crop}_AREA_1000_ha"
    prod = f"{crop}_PRODUCTION_1000_tons"
    yld = f"{crop}_YIELD_Kg_per_ha"

    valid = (df[area] > 0 ) & (df[prod] > 0 )

    df.loc[valid, yld] = (df.loc[valid, prod] * 1000) / df.loc[valid, area]

    return df
```

```python
crops = ["RICE","WHEAT","KHARIF_SORGHUM","RABI_SORGHUM","PEARL_MILLET","MAIZE","FINGER_MILLET","GROUNDNUT","SUNFLOWER","SOYABEAN","OILSEEDS","SUGARCANE","COTTON"]
for crop in crops:
    df = recalc_yield(df, crop)
```
